```python
import os
import json
import pandas as pd

# Define the directory containing the trial folders
directory = r"E:\chatbot-01\python-backend\trained-models\HRprediction_LSTM\HR_LSTM_Tuning\LSTM_Hyperparameter_Tuning"

# Initialize an empty list to store the extracted data
data = []

# Loop through each trial folder in the directory
for trial_folder in os.listdir(directory):
    trial_path = os.path.join(directory, trial_folder)
    
    # Check if the folder contains a "trial.json" file
    if os.path.isdir(trial_path) and 'trial.json' in os.listdir(trial_path):
        json_file_path = os.path.join(trial_path, 'trial.json')
        
        # Load the JSON data
        with open(json_file_path, 'r') as file:
            trial_data = json.load(file)
        
        # Extract the required values
        values = trial_data['hyperparameters']['values']
        metrics = trial_data['metrics']['metrics']
        score = trial_data.get('score', None)
        
        # Append the extracted values and metrics to the list
        data.append({
            'trial_id': trial_data['trial_id'],
            'lstm_units': values['lstm_units'],
            'dropout_rate': values['dropout_rate'],
            'dense_units': values['dense_units'],
            'learning_rate': values['learning_rate'],
            'accuracy': metrics['accuracy']['observations'][0]['value'][0],
            'loss': metrics['loss']['observations'][0]['value'][0],
            'val_accuracy': metrics['val_accuracy']['observations'][0]['value'][0],
            'val_loss': metrics['val_loss']['observations'][0]['value'][0],
            'score': score
        })

# Convert the data to a DataFrame for easier viewing
df = pd.DataFrame(data)

# Print the DataFrame as a matrix table
print(df.to_string(index=False))

# Define the path to export the DataFrame as a CSV file
output_csv_path = r"E:\chatbot-01\python-backend\trained-models\HRprediction_LSTM\HR_LSTM_Tuning\LSTM_Hyperparameter_Tuning\lstm_hyperparameter_tuning_results.csv"

# Export the DataFrame to a CSV file
df.to_csv(output_csv_path, index=False)

print(f"Results have been exported to {output_csv_path}")
```
